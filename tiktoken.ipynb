```python
import tiktoken
import pandas as pd
```

```python
df = pd.read_csv("ARES2_EJECUCION_ACTIVIDADES.csv")
df.head()
```

```python
df.describe
```

```python
descriptions = df['DESCRIPCION'].tolist()
descriptions
```

```python
encoding = tiktoken.get_encoding("cl100k_base")
```

```python
# Tokenize the text column and add a new column with the tokenized version
df["tokens"] = df["DESCRIPCION"].apply(lambda text: encoding.encode(text))

# Display the updated DataFrame
df.tail()
```

```python
df["text"] = df["tokens"].apply(lambda tokens: " ".join([encoding.decode_single_token_bytes(token).decode(errors="replace") for token in tokens]))
```

```python
df.tail()
```

```python
df
```

```python
df_sample = df.head(1000)
```

```python
unique_tokens = set(df_sample["tokens"].sum())
vocabulary_size = len(unique_tokens)
print(f"Vocabulary size: {vocabulary_size}")
```

```python
unique_words = df_sample["DESCRIPCION"].str.split().explode().nunique()
print(f"Number of unique words: {unique_words}")
```

```python
total_characters = df_sample["DESCRIPCION"].str.len().sum()
print(f"Total number of characters: {total_characters}")
```

```python
total_tokens = df_sample["tokens"].apply(len).sum()
print(f"Total number of tokens: {total_tokens}")
```

```python
word_frequency = df_sample["DESCRIPCION"].str.split().explode().value_counts()
print(word_frequency)
```

```python
token_frequency = df_sample["tokens"].explode().value_counts()
print(token_frequency)
```

```python
text_frequency = df_sample["text"].str.split().explode().value_counts()
print(word_frequency)
```

```python
N = 20
print("Word frequency:")
print(word_frequency.head(N))
print("\nToken frequency:")
print(text_frequency.head(N))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df_sample["DESCRIPCION"], df_sample["CODIGO_ETAPA"], test_size=0.2, random_state=42)

# Create the vectorizer and transform the training data
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)

# Train the logistic regression model
clf = LogisticRegression()
clf.fit(X_train_vec, y_train)

# Transform the test data and evaluate the model
X_test_vec = vectorizer.transform(X_test)
y_pred = clf.predict(X_test_vec)
accuracy = accuracy_score(y_test, y_pred)
print(f"Word-based classification accuracy: {accuracy}")
```
